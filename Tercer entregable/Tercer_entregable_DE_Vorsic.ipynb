## Se procede a realizar un proceso de ETL con librerÃ­a Panda

#### La API tiene una serie de limitaciones que comprometen el proceso que continÃºa: Por un lado, es informaciÃ³n que se encuentra discontinuada, no se actualiza. AdemÃ¡s, se encuentra limitada Ãºnicamente a un perÃ­odo breve. Por otro lado, la transformaciÃ³n inicial de los datos no fue realizada de manera exhaustiva. 
#### En base a estos puntos, se ha tomado la decisiÃ³n de extraer los datos desde la misma fuente en la que se han reservado para hacer una transformaciÃ³n sobre los mismos y dejar como resultado una base con un registro mÃ¡s funcional para futuras consultas. Se opta ademÃ¡s por reservar los registros originales dentro de la reserva de datos inicial y generar una base nueva con el formato ajustado. El cuidado de un registro original ayuda a preservar la integridad de los datos en caso de cualquier error o desatino en la base transformada.

```python
# Se extraen los datos desde la base de RedShift
query = f"select * from {env['AWS_REDSHIFT_SCHEMA']}.dfCovidColombia"
dfCol2 = spark.read \
    .format("jdbc") \
    .option("url", f"jdbc:postgresql://{env['AWS_REDSHIFT_HOST']}:{env['AWS_REDSHIFT_PORT']}/{env['AWS_REDSHIFT_DBNAME']}") \
    .option("dbtable", f"({query}) as tmp_table") \
    .option("user", env['AWS_REDSHIFT_USER']) \
    .option("password", env['AWS_REDSHIFT_PASSWORD']) \
    .option("driver", "org.postgresql.Driver") \
    .load()
```

```python
# Se confirma que el esquema sea correcto y que el proceso de guardado se haya realizado correctamente.
dfCol2.printSchema()
dfCol2.show()
```

```python
#ConversiÃ³n de Spark a Pandas para realizar las transformaciones
dfCol3 = dfCol2.toPandas()
print(dfCol3)
```

### **AnÃ¡lisis de las variables con datos nulos**

#### Se exploran los datos y su variedad dentro de las columnas que presentan datos nulos para determinar una acciÃ³n a tomar con las mismas. Se pude notar que una de las principales causantes de los datos nulos es la diferencia entre fallecidos y no fallecidos. Este es un dato relevante, por tanto, se evaluarÃ¡ la manera de que se pueda mantener la distinciÃ³n sin perjudicar la integridad de la base.

```python
dfCol3.tipo_recuperacion.value_counts()
```

```python
dfCol3.fecha_muerte.value_counts()
```

#### La columna fecha_muerte presenta demasiados valores nulos. Esto se debe a que la gran mayorÃ­a de los pacientes se ha recuperdado. La fecha resulta relevante para hacer estudios posteriores. Por tanto, se intentarÃ¡ conservar la columna para anÃ¡lisis posteriores. Los datos nulos serÃ¡n reemplazados por N/A (No aplica).

```python
dfCol3.fecha_recuperado.value_counts()
```

#### La columna fecha_recuperado tiene un pequeÃ±o nÃºmero de datos nulos. Lo ideal serÃ­a poder conservar la informaciÃ³n aplicando algÃºn mÃ©todo de sustituciÃ³n. Los datos nulos serÃ¡n reemplazados por N/A.

```python
dfCol3.tipo_recuperacion.value_counts()
```

#### La columna tipo_recuperacion podrÃ­a ser relevante para posteriores estudios. Se decide sustituir los valores nulos por N/A.

```python
dfCol3.nom_grupo_.value_counts()
```

#### La columna Nom_grupo_ no presenta una distinciÃ³n de relevancia para el estudio, por tanto se decide borrarla.

```python
# eliminacÃ³n de columna nom_grupo_
col_del=['nom_grupo_']
dfCovCol=dfCol3.drop(labels=col_del, axis=1)
dfCovCol.columns
```

```python
# SustituciÃ³n de valores en las restantes columnas a preservar.
col_sus= ['fecha_muerte','fecha_recuperado','tipo_recuperacion']
for i in col_sus:
  dfCovCol[i] = dfCovCol[i].replace({'NaN':'N/A'})
```

## VerificaciÃ³n de datos nulos

```python
dfCovCol.info()
```

#### No se presentan datos nulos, el proceso ha sido exitoso.

## Ajuste de tipos de datos

```python
# Se ajustan las columnas con variables numÃ©ricas
dfCovCol['id_de_caso'] = dfCovCol['id_de_caso'].astype('int')
dfCovCol['departamento'] = dfCovCol['departamento'].astype('int')
dfCovCol['edad'] = dfCovCol['edad'].astype('int')
dfCovCol['per_etn_'] = dfCovCol['per_etn_'].astype('int')
dfCovCol['ciudad_municipio'] = dfCovCol['ciudad_municipio'].astype('int')
```

```python
# Se ajustan las columnas con variables temporales
dfCovCol['fecha_reporte_web'] = dfCovCol['fecha_reporte_web'].astype('datetime64[ns]')
dfCovCol['fecha_de_notificaci_n'] = dfCovCol['fecha_de_notificaci_n'].astype('datetime64[ns]')
dfCovCol['fecha_inicio_sintomas'] = dfCovCol['fecha_inicio_sintomas'].astype('datetime64[ns]')
dfCovCol['fecha_diagnostico'] = dfCovCol['fecha_diagnostico'].astype('datetime64[ns]')
```

# Cargar datos actualizados en la tabla de Redshift

```python
# Connect to Redshift using psycopg2
conn = psycopg2.connect(
    host=env['AWS_REDSHIFT_HOST'],
    port=env['AWS_REDSHIFT_PORT'],
    dbname=env['AWS_REDSHIFT_DBNAME'],
    user=env['AWS_REDSHIFT_USER'],
    password=env['AWS_REDSHIFT_PASSWORD']
)
```

```python
cursor = conn.cursor()
cursor.execute(f"""
create table if not exists {env['AWS_REDSHIFT_SCHEMA']}.dfCovidColombiaConsolidado (
    "fecha_reporte_web" DATETIME,
    "id_de_caso" INT,
    "fecha_de_notificaci_n" DATETIME,
    "departamento" INT,
    "ciudad_municipio" INT,
    "edad" INT,
    "unidad_medida" VARCHAR(50),
    "sexo" VARCHAR(50),
    "fuente_tipo_contagio" VARCHAR(50),
    "ubicacion" VARCHAR(50),
    "estado" VARCHAR(50),
    "recuperado" VARCHAR(50),
    "fecha_inicio_sintomas" DATETIME,
    "fecha_diagnostico" DATETIME,
    "fecha_recuperado" VARCHAR(50),
    "tipo_recuperacion" VARCHAR(50),
    "per_etn_" INT,
    "fecha_muerte" VARCHAR(50)
) sortkey(id_de_caso);
""")
conn.commit()
cursor.close()
print("Table created!")
```
